5장.다중 선형 회귀

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6, 8])                # 공부 시간 
x2 = np.array([0, 4, 2, 3])                # 과외 시간
y = np.array([81, 93, 91, 97])             # 성적

# 데이터의 분포 그래프로 나타내기
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter3D(x1, x2, y);
plt.show()

a1 = 0            # 기울기
a2 = 0
b = 0             # 절편

# 학습률
lr = 0.01

# 몇 번 반복될지 설정
epochs = 2001

# x값의 개수
n = len(x1)

# 경사 하강법
for i in range(epochs):
  y_pred = a1 * x1 + a2 * x2 + b       # 예측 값을 구하는 식
  error = y - y_pred                   # 실제값과 비교한 오차

  # 오차 함수를 편미분
  a1_diff = (2/n) * sum(-x1 * (error))
  a2_diff = (2/n) * sum(-x2 * (error))
  b_diff = (2/n) * sum(-(error))

  # 학습률을 곱해 기존의 값 업데이트
  a1 = a1 - lr * a1_diff
  a2 = a2 - lr * a2_diff
  b = b - lr * b_diff

  if i % 100 == 0:
    print("epochs = %.f, 기울기1 = %.04f, 기울기2 = %.04f, 절편 = %.04f" % (i, a1, a2, b), '\n')

  print("실제 점수 : ", y)
  print("예측 점수 : ", y_pred)
```
