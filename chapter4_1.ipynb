4장.최소 제곱

```python
import numpy as np                # 넘파이 라이브러리를 np라는 이름으로 불러옴

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

mx = np.mean(x)                                   # x의 평균값
my = np.mean(y)                                   # y의 평균값

print("x의 평균값 : ", mx)
print("y의 평균값 : ", my, '\n')

divisor = sum([(i - mx) ** 2 for i in x])            # 기울기 공식의 분모

# 기울기 공식의 분자
def top(x, mx, y, my):
  d = 0;                                             # 초깃값을 0으로 설정
  for i in range(len(x)):                            # x 개수(4)만큼 실행
    d += (x[i] - mx) * (y[i] - my)                   # 최소 제곱법 구현
  return d
dividend = top(x, mx, y, my)

print("분모 : ", divisor)
print("분자 : ", dividend, '\n')

a = dividend / divisor                            # 기울기 a  

b = my - (mx * a)                                 # y절편 b

print("기울기 a = ", a)
print("y 절편 b = ", b)
```
