4장.평균 제곱 오차

```python
import numpy as np

# 가상의 기울기가 a와 y절편 b를 정의
fake_a = 3
fake_b = 76

# 공부 시간 x와 성적 y의 넘파이 배열 생성
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# y = ax + b에 가상의 a, b 값을 대입한 결과 출력하는 함수 정의
def predict(x):                         
  return fake_a * x + fake_b                          

# 결과값이 들어갈 빈 리스트 생성
predict_result = []

# 모든 x값을 한 번씩 대입해 prdict_result 리스트 완성
for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("공부시간 = %.f, 실제점수 = %.f, 예측점수 = %.f" %(x[i], y[i], predict(x[i])))

# 평균 제곱 오차를 구하는 함수식 정의
n = len(x)
def mse(y, y_pred):
  return (1/n) * sum((y - y_pred) ** 2)

print("평균 제곱 오차 : " + str(mse(y, predict_result)))
```
